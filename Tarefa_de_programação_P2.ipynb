# 1 - Calculadora de ângulo entre ponteiros do relógio

```python
def calcular_angulo(hora, minuto):
    if hora < 0 or hora >= 24 or minuto < 0 or minuto >= 60:
        raise ValueError("Horário inválido")
    angulo_hora = (hora * 30)
    angulo_minuto = (minuto * 6)
    angulo_final = abs(angulo_hora - angulo_minuto)
    menor_angulo = min(angulo_final, 360 - angulo_final)
    return (menor_angulo)
```

```python
def main():
    while True:
        entrada = input("Digite o horário no formato 'hh:mm' ou digite 'f' para encerrar o programa): ")
        if entrada.lower() == 'f':
            print("Programa encerrado.")
            break

        try:
            hora, minuto = map(int, entrada.split(':'))
            angulo = abs(calcular_angulo(hora, minuto))
            print(f"O menor ângulo é de: {angulo} °")
        except ValueError as e:
            print(f"Erro: {e}. Input inválido.")
main()
```

```python
def validar_movimento_cavalo(posicao_inicial, posicao_final):
    movimentos_cavalo = [(2, 1), (2, -1), (-2, 1), (-2, -1),(1, 2), (1, -2), (-1, 2), (-1, -2)]

    if len(posicao_inicial) != 2 or len(posicao_final) != 2:
        raise ValueError("Formato de posição inválido")

    posicao_inicial = posicao_inicial.lower()
    posicao_final = posicao_final.lower()

    colunas = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5, 'f': 6, 'g': 7, 'h': 8}
    linha_inicial = int(posicao_inicial[1])
    coluna_inicial = colunas[posicao_inicial[0]]
    linha_final = int(posicao_final[1])
    coluna_final = colunas[posicao_final[0]]

    if linha_final < 1 or linha_final > 8 or coluna_final < 1 or coluna_final > 8:
        raise ValueError("Posição final fora do tabuleiro")

    diff_linha = abs(linha_final - linha_inicial)
    diff_coluna = abs(coluna_final - coluna_inicial)
    if (diff_linha, diff_coluna) in movimentos_cavalo:
        return True
    else:
        return False
```

```python
def main():
    while True:
        entrada = input("Digite a posição inicial e a posição final do cavalo (exemplo de como escrever: 'd4 e f5') "
                        "ou 'f' para sair: ")
        if entrada.lower() == 'f':
            print("Encerrando o programa.")
            break

        try:
            posicoes = entrada.split("e")
            posicao_inicial = posicoes[0].strip()
            posicao_final = posicoes[1].strip()
            if validar_movimento_cavalo(posicao_inicial, posicao_final):
                print("VÁLIDO.")
            else:
                print("INVÁLIDO.")
        except ValueError as e:
            print(f"Erro: {e}. Insira posições válidas.")

main()
```

# 3 - Calculadora de troco



```python
def calcular_notas_moedas(valor):
    notas = [100, 50, 20, 10, 5, 2]
    moedas = [1, 0.50, 0.25, 0.10, 0.05, 0.01]

    contador_notas = {nota: 0 for nota in notas}
    contador_moedas = {moeda: 0 for moeda in moedas}

    for nota in notas:
        contador_notas[nota] = valor // nota
        valor %= nota

    for moeda in moedas:
        contador_moedas[moeda] = int(valor / moeda)
        valor %= moeda
        valor = round(valor, 2)

    return contador_notas, contador_moedas
```

```python
def main():
    while True:
        try:
            valor = float(input("Digite o valor monetário (com duas casas decimais): "))
            if valor <= 0:
                raise ValueError("O valor deve ser maior que zero")

            notas, moedas = calcular_notas_moedas(valor)

            print("NOTAS:")
            for nota, quantidade in notas.items():
                print(f"{quantidade} nota(s) de R$ {nota:.2f}")

            print("MOEDAS:")
            for moeda, quantidade in moedas.items():
                print(f"{quantidade} moeda(s) de R$ {moeda:.2f}")

            break
        except ValueError as e:
            print(f"Erro: {e}. Por favor, insira um valor monetário válido.")


main()
```

# 4 - Frequência de números

```python
def calcular_frequencia(numeros):
    frequencia = {}
    for numero in numeros:
        if numero in frequencia:
            frequencia[numero] += 1
        else:
            frequencia[numero] = 1
    return frequencia
```

```python
def main():
    numeros = []
    while True:
        entrada = input("Digite um número inteiro ou 'f' para encerrar o programa: ")
        if entrada.lower() == 'f':
            break

        try:
            numero = int(entrada)
            numeros.append(numero)
        except ValueError:
            print("Digite um número inteiro.")

    frequencia = calcular_frequencia(numeros)

    for numero, freq in frequencia.items():
        print(f"O número {numero} apareceu {freq} vezes.")
main()
```
